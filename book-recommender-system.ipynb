```python
rating_avg_df = rating_new.groupby('Book-Title')['Book-Rating'].mean().reset_index()
rating_avg_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
rating_avg_df
```

```python
popular_df=rating_no_df.merge(rating_avg_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')
```

```python
popular_df=popular_df[['Book-Title','Book-Author','num_rating','avg_rating','Image-URL-M']]
```

```python
final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score=cosine_similarity(pt)
```

```python
def recommend(book_name):
    book_index=np.where(pt.index==book_name)[0][0]
    similar_book=sorted(list(enumerate(similarity_score[book_index])),key=(lambda x:x[1]),reverse=True)[1:5]
    data=[]
    for i in similar_book:
        item=[]
        temp = books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
        
        
    return data
```

```python
recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```
